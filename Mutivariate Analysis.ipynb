#### Multivariate Analysis

```python
import pandas as pd

seats_df = pd.read_csv('seats_reshaped_interpolated_clean.csv')
gii_df = pd.read_csv('df_gii_filtered_cleaned.csv')
female_labor_df = pd.read_csv('df_female_labor.csv')

seats_df
```

```python
gii_df['Country Code'] = gii_df['iso3']
gii_df['Country Name'] = gii_df['country']
gii_df.drop(['iso3', 'country'], axis=1, inplace=True)
gii_df
```

```python
gii_long = gii_df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='GII')
labor_long = female_labor_df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Labor Force')
seats_long = seats_df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Seats by Women')

gii_long['Year'] = gii_long['Year'].astype(int)
labor_long['Year'] = labor_long['Year'].astype(int)
seats_long['Year'] = seats_long['Year'].astype(int)

merged = gii_long.merge(labor_long, on=['Year', 'Country Name', 'Country Code'], how='inner') \
                 .merge(seats_long, on=['Year', 'Country Name', 'Country Code'], how='inner')

merged.set_index(['Year', 'Country Name', 'Country Code'], inplace=True)

merged
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

variables = ['GII', 'Labor Force', 'Seats by Women']

for var in variables:
    plt.figure(figsize=(8, 6))
    sns.histplot(merged[var], kde=True, bins=30)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()

skewness = merged[variables].skew()
print("Skewness of Variables:")
print(skewness)
```

```python
import numpy as np
merged['Log Seats by Women'] = np.log1p(merged['Seats by Women'])

plt.figure(figsize=(8, 6))
sns.histplot(merged['Log Seats by Women'], kde=True, bins=30)
plt.title('Histogram of Log Transformed Seats by Women')
plt.xlabel('Log Seats by Women')
plt.ylabel('Frequency')
plt.show()

log_skewness = merged['Log Seats by Women'].skew()
print(f"Skewness of Log Transformed Seats by Women: {log_skewness}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(24, 6), sharey=False)

sns.scatterplot(ax=axes[0], x=merged['GII'], y=merged['Labor Force'])
axes[0].set_title('GII vs Labor Force')
axes[0].set_xlabel('GII')
axes[0].set_ylabel('Labor Force')
axes[0].grid(True)

sns.scatterplot(ax=axes[1], x=merged['Log Seats by Women'], y=merged['Labor Force'])
axes[1].set_title('Log Seats by Women vs Labor Force')
axes[1].set_xlabel('Log Seats by Women')
axes[1].grid(True)

sns.scatterplot(ax=axes[2], x=merged['GII'], y=merged['Log Seats by Women'])
axes[2].set_title('GII vs Log Seats by Women')
axes[2].set_xlabel('GII')
axes[2].set_ylabel('Log Seats by Women')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import pearsonr

variables = ['GII', 'Log Seats by Women', 'Labor Force']
results = []

for i, var1 in enumerate(variables):
    for j, var2 in enumerate(variables):
        if i < j:  
            corr, p_value = pearsonr(merged[var1], merged[var2])
            results.append((var1, var2, corr, p_value))

            if abs(corr) >= 0.7:
                strength = "strong"
            elif abs(corr) >= 0.4:
                strength = "moderate"
            else:
                strength = "weak"

            print(f"Pearson Correlation between {var1} and {var2}:")
            print(f"  Correlation Coefficient: {corr:.2f}")
            print(f"  P-value: {p_value:.4f}")
            print(f"  Strength of Correlation: {strength}")
            print("")
```

```python
from scipy.stats import ttest_ind

for var in ['GII', 'Log Seats by Women']:
    median_value = merged[var].median()
    group_low = merged[merged[var] <= median_value]
    group_high = merged[merged[var] > median_value]

    stat, p_value = ttest_ind(group_low['Labor Force'], group_high['Labor Force'], equal_var=False)
    
    print(f"\nHypothesis Test for {var}:")
    print(f"  Median Value: {median_value}")
    print(f"  T-test Statistic: {stat}")
    print(f"  P-value: {p_value}")
    if p_value < 0.05:
        print(f"  Reject the null hypothesis: Significant difference in Labor Force between groups.")
    else:
        print(f"  Fail to reject the null hypothesis: No significant difference in Labor Force between groups.")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Initialize an empty list to store the results
results = []

for var in ['GII', 'Log Seats by Women']:
    median_value = merged[var].median()
    group_low = merged[merged[var] <= median_value]
    group_high = merged[merged[var] > median_value]

    stat, p_value = ttest_ind(group_low['Labor Force'], group_high['Labor Force'], equal_var=False)

    # Append the results to the list
    results.append({
        'Variable': var,
        'Median Value': median_value,
        'T-test Statistic': stat,
        'P-value': p_value,
        'Significance': 'Significant' if p_value < 0.05 else 'Not Significant'
    })

# Convert the results list into a DataFrame
results_df = pd.DataFrame(results)

# Print the results table
print("\nHypothesis Test Results:")
results_df
```

```python
import statsmodels.api as sm

X = merged[['GII', 'Log Seats by Women']]
y = merged['Labor Force'] 

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

merged['Residuals'] = model.resid
```

```python
import matplotlib.pyplot as plt

sm.qqplot(merged['Residuals'], line='s')
plt.title('QQ Plot of Residuals')
plt.show()
```

```python
from linearmodels.panel import PanelOLS
from statsmodels.tools import add_constant  # Import add_constant from statsmodels
import pandas as pd

# Reset the index to access Year and Country Code as columns
merged = merged.reset_index()

# Set the multi-index for panel data
merged = merged.set_index(['Country Code', 'Year'])  # Use Country Code and Year as multi-index

# Define dependent and independent variables
y = merged['Labor Force']  # Dependent variable
X = merged[['GII', 'Seats by Women']]  # Independent variables
X = add_constant(X)  # Add a constant term (intercept)

# Fit the fixed-effects model
model = PanelOLS(y, X, entity_effects=True)  # entity_effects=True specifies fixed effects
results = model.fit()

# Display the results
print(results.summary)
```
